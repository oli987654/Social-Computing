```python
## your code here
import time
import selenium
import threading
from selenium import webdriver
from selenium.webdriver import Chrome
from selenium.webdriver.common.by import By


browsers = []
browsers.append([webdriver.Edge(options=webdriver.EdgeOptions()),"login"])
browsers.append([Chrome(),"login"])
browsers.append([webdriver.Edge(options=webdriver.EdgeOptions()),""])
browsers.append([Chrome(),""])


def getBrowser(aBrowserList):
   aBrowser=aBrowserList[0]
   aBrowser.maximize_window()
   aBrowser.get("https://www.trivago.ch")
   time.sleep(3)

   #Go to login page
   if (aBrowserList[1]=="login"):
      elements = aBrowser.find_elements(By.CSS_SELECTOR, "span.flex-grow.text-m.mx-2.whitespace-nowrap")
      for element in elements:
         if "Einloggen" in element.text:
            element.click()
            break
      time.sleep(2)
      # click on email field, enter address and hit log-in button
      email_field = aBrowser.find_element(By.CLASS_NAME, "BaseInput_input__0WXAa")
      time.sleep(2)
      email_field.click()
      email_field.send_keys("oli987654@gmx.ch")
      next_button = aBrowser.find_element(By.XPATH, "//button[@data-testid='login-next-button']").click()
      time.sleep(2)
      # the same with password
      password_field = aBrowser.find_element(By.CLASS_NAME, "PasswordStrengthInput_input__AcvDF")
      password_field.click()
      password_field.send_keys("aPasswordForTrivago123")
      aBrowser.find_element(By.XPATH, "//button[@data-testid='login-submit']").click()
      time.sleep(2)

   #Search for Hotels in Berlin
   search_field = aBrowser.find_element(By.CLASS_NAME, "SearchInput_input__mPUpa")
   search_field.click()
   search_field.send_keys("Berlin")
   time.sleep(1)
   aBrowser.find_element(By.CSS_SELECTOR, "button[data-testid='search-button']").click()
   #We have to wait 35 seconds because Trivago checks if you are a computer or automated system LOL
   time.sleep(35)

   #Time to look at the website
   time.sleep(120)

   aBrowser.quit()
   

threads = []
for i, browser in enumerate(browsers):
   thread = threading.Thread(target=getBrowser, args=[browser])
   threads.append(thread)
   thread.start()

for thread in threads:
   thread.join()

```
